```python
import json
import pyecharts.options as opts
from pyecharts.charts import Radar
from pyecharts.charts import Line
from pyecharts.commons.utils import JsCode
```

```python
from pyecharts.globals import CurrentConfig
CurrentConfig.ONLINE_HOST = "http://127.0.0.1:8000/assets/"
```

```python
# 背景色
background_color_js = (
    "new echarts.graphic.LinearGradient(0, 0, 0, 1, "
    "[{offset: 0, color: '#5c5959'}, {offset: 1, color: '#909090'}], false)"
)
```

```python
# CNN4模型
for i in range(0,20):
    f = open("./data/cnn4/aapd/metrics_epoch_"+str(i)+".json", encoding='gbk')
    data = json.load(f)
    Train = [[data['training_HL'],data['training_R_at_1'],data['training_R_at_3'],
         data['training_R_at_5'],data['training_P_at_1'],data['training_P_at_3'],
         data['training_P_at_5'],data['training_FBeta_at_1'],data['training_FBeta_at_3'],
         data['training_FBeta_at_5'],data['training_nDCG_at_1'],data['training_nDCG_at_3'],
         data['training_nDCG_at_5'],data['training_macro_f1'],data['training_micro_precision'],
         data['training_micro_recall'],data['training_micro_fscore'],data['training_loss']]]
    Validation = [[data['validation_HL'],data['validation_R_at_1'],data['validation_R_at_3'],
         data['validation_R_at_5'],data['validation_P_at_1'],data['validation_P_at_3'],
         data['validation_P_at_5'],data['validation_FBeta_at_1'],data['validation_FBeta_at_3'],
         data['validation_FBeta_at_5'],data['validation_nDCG_at_1'],data['validation_nDCG_at_3'],
         data['validation_nDCG_at_5'],data['validation_macro_f1'],data['validation_micro_precision'],
         data['validation_micro_recall'],data['validation_micro_fscore'],data['validation_loss']]]
    Best_Validation = [[data['best_validation_HL'],data['best_validation_R_at_1'],data['best_validation_R_at_3'],
         data['best_validation_R_at_5'],data['best_validation_P_at_1'],data['best_validation_P_at_3'],
         data['best_validation_P_at_5'],data['best_validation_FBeta_at_1'],data['best_validation_FBeta_at_3'],
         data['best_validation_FBeta_at_5'],data['best_validation_nDCG_at_1'],data['best_validation_nDCG_at_3'],
         data['best_validation_nDCG_at_5'],data['best_validation_macro_f1'],data['best_validation_micro_precision'],
         data['best_validation_micro_recall'],data['best_validation_micro_fscore'],data['best_validation_loss']]]
    (
        Radar(init_opts=opts.InitOpts(width="1280px", height="720px", bg_color=JsCode(background_color_js)))
        .add_schema(
            schema=[
                opts.RadarIndicatorItem(name="1-10×HL", max_=1),
                opts.RadarIndicatorItem(name="R@1", max_=1),
                opts.RadarIndicatorItem(name="R@3", max_=1),
                opts.RadarIndicatorItem(name="R@5", max_=1),
                opts.RadarIndicatorItem(name="P@1", max_=1),
                opts.RadarIndicatorItem(name="P@3", max_=1),
                opts.RadarIndicatorItem(name="P@5", max_=1),
                opts.RadarIndicatorItem(name="F1@1", max_=1),
                opts.RadarIndicatorItem(name="F1@3", max_=1),
                opts.RadarIndicatorItem(name="F1@5", max_=1),
                opts.RadarIndicatorItem(name="nDCG@1", max_=1),
                opts.RadarIndicatorItem(name="nDCG@3", max_=1),
                opts.RadarIndicatorItem(name="nDCG@5", max_=1),
                opts.RadarIndicatorItem(name="MaF1", max_=1),
                opts.RadarIndicatorItem(name="MiP", max_=1),
                opts.RadarIndicatorItem(name="MiR", max_=1),
                opts.RadarIndicatorItem(name="MiF1", max_=1),
                opts.RadarIndicatorItem(name="loss", max_=1),
            ],
            splitarea_opt=opts.SplitAreaOpts(
                is_show=True, areastyle_opts=opts.AreaStyleOpts(opacity=1)
            ),
            textstyle_opts=opts.TextStyleOpts(color="#FFFFFF"),
        )
        .add(
            series_name="Train",
            data=Train,
            linestyle_opts=opts.LineStyleOpts(opacity=0.5, color="#00FFFF"),
            areastyle_opts=opts.AreaStyleOpts(opacity=0.3, color="#00FFFF"),
            label_opts=opts.LabelOpts(is_show=False,color="#00FFFF"),
        )
        .add(
            series_name="Validation",
            data=Validation,
            linestyle_opts=opts.LineStyleOpts(opacity=0.5, color="#00FF00"),
            areastyle_opts=opts.AreaStyleOpts(opacity=0.3, color="#00FF00"),
            label_opts=opts.LabelOpts(is_show=False,color="#00FF00"),
        )
        .add(
            series_name="Best_Validation",
            data=Best_Validation,
            linestyle_opts=opts.LineStyleOpts(opacity=0.5, color="#00008B"),
            areastyle_opts=opts.AreaStyleOpts(opacity=0.2, color="#00008B"),
            label_opts=opts.LabelOpts(is_show=False,color="#00008B"),
        )
#         .set_series_opts(label_opts=opts.LabelOpts(is_show=False,color='auto'))
        .set_global_opts(
            title_opts=opts.TitleOpts(title="CNN4 Metrics Epoch "+str(i),
                                     title_textstyle_opts=opts.TextStyleOpts(color="#e3dfde")), 
            legend_opts=opts.LegendOpts(textstyle_opts=opts.TextStyleOpts(color="#FFFFFF"),inactive_color="#c1c1be")
        )
        .render("cnn4_epoch_"+str(i)+".html")
    )
```

```python
# 模型最佳对比
f1 = open("./data/cnn4/aapd/metrics.json", encoding='gbk')
f2 = open("./data/lsan4/aapd/metrics.json", encoding='gbk')
data1 = json.load(f1)
data2 = json.load(f2)
test_cnn4 = [[data1['test_R_at_1'],data1['test_R_at_3'],
    data1['test_R_at_5'],data1['test_P_at_1'],data1['test_P_at_3'],
    data1['test_P_at_5'],data1['test_FBeta_at_1'],data1['test_FBeta_at_3'],
    data1['test_FBeta_at_5'],data1['test_nDCG_at_1'],data1['test_nDCG_at_3'],
    data1['test_nDCG_at_5'],data1['test_loss']]]
test_lsan4 = [[data2['test_R_at_1'],data2['test_R_at_3'],
    data2['test_R_at_5'],data2['test_P_at_1'],data2['test_P_at_3'],
    data2['test_P_at_5'],data2['test_FBeta_at_1'],data2['test_FBeta_at_3'],
    data2['test_FBeta_at_5'],data2['test_nDCG_at_1'],data2['test_nDCG_at_3'],
    data2['test_nDCG_at_5'],data2['test_loss']]]
(
    Radar(init_opts=opts.InitOpts(width="1280px", height="720px", bg_color=JsCode(background_color_js)))
    .add_schema(
        schema=[
            opts.RadarIndicatorItem(name="R@1", max_=1),
            opts.RadarIndicatorItem(name="R@3", max_=1),
            opts.RadarIndicatorItem(name="R@5", max_=1),
            opts.RadarIndicatorItem(name="P@1", max_=1),
            opts.RadarIndicatorItem(name="P@3", max_=1),
            opts.RadarIndicatorItem(name="P@5", max_=1),
            opts.RadarIndicatorItem(name="F1@1", max_=1),
            opts.RadarIndicatorItem(name="F1@3", max_=1),
            opts.RadarIndicatorItem(name="F1@5", max_=1),
            opts.RadarIndicatorItem(name="nDCG@1", max_=1),
            opts.RadarIndicatorItem(name="nDCG@3", max_=1),
            opts.RadarIndicatorItem(name="nDCG@5", max_=1),
            opts.RadarIndicatorItem(name="loss", max_=1),
        ],
        splitarea_opt=opts.SplitAreaOpts(
            is_show=True, areastyle_opts=opts.AreaStyleOpts(opacity=1)
        ),
        textstyle_opts=opts.TextStyleOpts(color="#FFFFFF"),
    )
    .add(
        series_name="cnn4",
        data=test_cnn4,
        linestyle_opts=opts.LineStyleOpts(opacity=0.5, color="#00008B"),
        areastyle_opts=opts.AreaStyleOpts(opacity=0.3, color="#00008B"),
        label_opts=opts.LabelOpts(is_show=False,color="#00008B"),
    )
    .add(
        series_name="lsan4",
        data=test_lsan4,
        linestyle_opts=opts.LineStyleOpts(opacity=0.5, color="#00FF00"),
        areastyle_opts=opts.AreaStyleOpts(opacity=0.3, color="#00FF00"),
        label_opts=opts.LabelOpts(is_show=False,color="#00FF00"),
    )
#         .set_series_opts(label_opts=opts.LabelOpts(is_show=False,color='auto'))
    .set_global_opts(
        title_opts=opts.TitleOpts(title="Comparison between cnn4 and lsan4",
                                title_textstyle_opts=opts.TextStyleOpts(color="#e3dfde")), 
        legend_opts=opts.LegendOpts(textstyle_opts=opts.TextStyleOpts(color="#FFFFFF"),inactive_color="#c1c1be")
    )
    .render("Comparison.html")
)
```

```python
# 模型曲线图
# cnn4模型
mif1=[]
for i in range(0,20):
    f = open("./data/cnn4/aapd/metrics_epoch_"+str(i)+".json", encoding='gbk')
    data = json.load(f)
    mif1.append(data['best_validation_micro_fscore'])
    
(
    Line(init_opts=opts.InitOpts(width="1280px", height="720px", bg_color=JsCode(background_color_js)))
    .set_global_opts(
        tooltip_opts=opts.TooltipOpts(is_show=False),
        xaxis_opts=opts.AxisOpts(type_="category",
                                 name_textstyle_opts=opts.TextStyleOpts(color="#FFFFFF"),
                                axislabel_opts=opts.LabelOpts(is_show=True,color="#FFFFFF"),
                                axisline_opts=opts.AxisLineOpts(linestyle_opts=opts.LineStyleOpts(color="#FFFFFF"))),
        yaxis_opts=opts.AxisOpts(
            type_="value",
            axistick_opts=opts.AxisTickOpts(is_show=True),
            splitline_opts=opts.SplitLineOpts(is_show=True),
            min_=0.55,
            name_textstyle_opts=opts.TextStyleOpts(color="#FFFFFF"),
            axislabel_opts=opts.LabelOpts(is_show=True,color="#FFFFFF"),
            axisline_opts=opts.AxisLineOpts(linestyle_opts=opts.LineStyleOpts(color="#FFFFFF")),
        ),
        legend_opts=opts.LegendOpts(textstyle_opts=opts.TextStyleOpts(color="#FFFFFF"),
                                    legend_icon="circle",inactive_color="#c1c1be")
    )
    .add_xaxis(xaxis_data=list(range(0,20)))
    .add_yaxis(
        series_name="Mif1",
        y_axis=mif1,
        symbol="circle",
        symbol_size=6,
        is_symbol_show=True,
        is_smooth=True,
        label_opts=opts.LabelOpts(is_show=False),
        linestyle_opts=opts.LineStyleOpts(width=2,color="#FFFFFF"),
        itemstyle_opts=opts.ItemStyleOpts(
            color="red", border_color="#fff", border_width=3
        ),
    )
    .render("cnn4_Mif1.html")
)
```
